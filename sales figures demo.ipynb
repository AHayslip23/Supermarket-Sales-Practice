```python
#simple linear regression
#simple linearRegression between quantity and total sales 
```

```python
simple1 = sales[["Quantity", "Total"]] 
```

```python
simple1
```

```python
X = simple1.iloc[:, :-1].values
Y = simple1.iloc[:, -1].values 
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
Y_train 
```

```python
Y_test 
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred 
```

```python
plt.scatter(X_train, Y_train, color="red")
plt.plot(X_train, regressor.predict(X_train), color="blue")
plt.title("Correalation between Quantity of Items and Total Sales")
plt.xlabel("Quantity of Items")
plt.ylabel("Total Sales")
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test, y_pred)
```

```python
#the model returns a correalation of 50.8% 
#there is a 50.8% correalation between Quantity of Items purchased and Total Sales 
```

```python
class1 = sales[["Quantity", "Unit price", "Tax 5%", "Total", "cogs", "gross margin percentage", "gross income", "Rating", "Credit Payment"]]
```

```python
class1
```

```python
X = class1.iloc[:, :-1].values
Y = class1.iloc[:, -1].values 
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
Y_train 
```

```python
Y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, Y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), Y_test.reshape(len(Y_test), 1)), 1)) 
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, y_pred)
print(cm)
accuracy_score(Y_test, y_pred)
```
